### <미션 1>
* 네이버 영화에서 랭킹 상위 50위의 영화 제목 수집

```python
import requests
from bs4 import BeautifulSoup

#get
URL = f"https://movie.naver.com/movie/sdb/rank/rmovie.naver"

headers={
    'referer': 'https://movie.naver.com/',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.99 Safari/537.36'
}

req = requests.get(URL,headers)
soup = BeautifulSoup(req.text, 'html.parser')

movies = soup.select('table>tbody>tr>td>div>a')

i = 1
for movie_name in movies:
    print(f'{i:2d}위: ',movie_name.get_text())
    i=i+1
```

### <미션 2> 
* 수집한 네이버 영화에서 장르(개요), 개봉 일자, 감독 수집

```python
HOME_URL = f"https://movie.naver.com"

movie_urls = soup.find("tbody")
movie_url_lst = []
i=0
rechar='\n\r\t'

for url in movie_urls.find_all("tr"):
    movie_url = url.find('td', {'class':'title'})
    if movie_url == None:
        continue
    movie_url_lst.append(HOME_URL + movie_url.find("a")['href'])

for MOVIE_INFO_URL in movie_url_lst:
    print(f'{i+1:2d}위: ',movies[i].get_text())
    i=i+1
    headers_tmp={
    'referer': 'https://movie.naver.com/movie/sdb/rank/rmovie.naver',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.99 Safari/537.36'
    }
    req_tmp = requests.get(MOVIE_INFO_URL,headers_tmp)
    soup_tmp = BeautifulSoup(req_tmp.text, 'html.parser')
    movie_info_data = soup_tmp.find('dl', {'class':'info_spec'})
    
    movie_infos = movie_info_data.select('dd>p>span')
    movie_info_lst = []
    for movie_info in movie_infos:
        movie_info_lst.append(movie_info.get_text().replace('\t','').replace('\r','').replace('\n','').strip())
    
    movie_director_infos = movie_info_data.select('dd>p>a')
    print(f"       {movie_info_lst[0]} | {movie_info_lst[-1]} | {movie_director_infos[0].get_text()}")
    
    print('\n')
```

### <미션3>
* 한국산업기술대학교 eclass 의 교육현황 탭에서 1페이지의 과목/교수/청강 가능 여부를 크롤링해서 정리후 출력해보세요.

```python
import requests
from bs4 import BeautifulSoup

url = "http://eclass.kpu.ac.kr/ilos/st/main/course_ing_list.acl"
headers={
    'referer': 'http://eclass.kpu.ac.kr/ilos/main/main_form.acl',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.99 Safari/537.36'
}
req = requests.get(url, headers=headers)
soup = BeautifulSoup(req.text, "html.parser")

info_data = soup.select('#content_text > table > tbody > tr ')
info_lst = []

for info in info_data:
    info_lst.append(info.get_text().strip())
    info_lst[-1] = info_lst[-1].split('\n')
    print(f"과목: {info_lst[-1][2][:-3]} \n교수: {info_lst[-1][2][-3:]}\n청강 가능 여부: {info_lst[-1][-1]}\n")
```
